```python
import matplotlib.pyplot as plt
import numpy as np
from sensitivities.distributions import Discrete, Gaussian, Uniform, Sinusoidal
```

## Distributions

Variuous individual distributions sampled for visualization

```python
n_hist = 200
n_samples = 100000

plt.figure(figsize=(9,4))
plt.hist(np.array(Sinusoidal(1,0.7).sample(n_samples)), n_hist)
plt.hist(np.array(Gaussian(1, 0.05).sample(n_samples)), n_hist)
plt.hist(np.array(Uniform(-0.4, 0.4).sample(n_samples)), n_hist)
plt.hist(np.array(Discrete([0.5, 1.2]).sample(n_samples//10)), n_hist)
plt.show()
```
